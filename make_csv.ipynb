```python
valid_files = tqdm(valid_files)
df = []
for file in valid_files:
  open_data = open("./train/"+valid_directories[0]+'/'+file).read()
  data = json.loads(open_data)
  ds = {}
  ds['id'] = file[:5]
  for a in range(2):
    for b in range(len(data['data'][a]['x'])):
      ds['x_1']=data['data'][a]['x'][b]
  df.append(ds)

df = pd.DataFrame(df)
df
```
